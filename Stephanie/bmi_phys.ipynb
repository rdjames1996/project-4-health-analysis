```python
# Import dependencies
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in csv file
file_path = pd.read_csv(f"/Users/stephanietran/Desktop/project 4 play/diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
file_path
```

```python
# Convert csv to dataframe
bmi_df = pd.DataFrame(file_path)
bmi_df
```

```python
# List columns for review
bmi_df.columns.to_list()
```

```python
# Drop unnecessary columns
bmi_df_drop = bmi_df.drop(columns=['NoDocbcCost', 'GenHlth', 'MentHlth', 'PhysHlth', 'DiffWalk'])
bmi_df_drop
```

```python
bmi_df_drop.groupby("Diabetes_binary").count()
```

```python
# Filter through df to pull needed columns and rename 'Diabetes_binary ' to 'Diabetes_status' to pull target variable and features
phys_df = bmi_df.filter(['Diabetes_binary', 'BMI', 'PhysActivity'], axis=1)
phys_df.rename(columns={'Diabetes_binary': 'Diabetes_Status'}, inplace=True)
phys_df
```

```python
# Convert new dataframe to csv file for future refernces
phys_df.to_csv("/Users/stephanietran/Desktop/project-4-health-analysis/Resources/phys_df.csv")
```

```python
# Check df and data types  
phys_df.dtypes
```

```python
# Split data into target variable and features
X = phys_df.drop(columns=['Diabetes_Status'], axis=1)
y = phys_df['Diabetes_Status']
```

```python
X 
```

## Split data into training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state=1,
                                                    stratify=y)
```

## Create a Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

## Fit (train) or Model using the training data

```python
# Train the data
classifier.fit(X_train, y_train)
```

## Validating the model using the test data

```python
# Score the data
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

## Make Predictions

```python
# Predict outcomes for test data set
predictions =classifier.predict(X_test)
predictions_df = pd.DataFrame({"Prediction": predictions, "Actual": y_test})
predictions_df
```

```python
predictions
```

```python
y_test
```

## Calculate the Accuracy Score

```python
from sklearn.metrics import accuracy_score
# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```

## Confusion Matrix & Classification Report

```python
from sklearn.metrics import confusion_matrix
confusion_mat=confusion_matrix(y_test, predictions)
confusion_mat
```

```python
from sklearn.metrics import classification_report
target_names = ["0", "1"]
print(classification_report(y_test, predictions, target_names=target_names))
```

```python
fig, ax = plt.subplots()
cax = ax.matshow(confusion_mat, cmap=plt.cm.Blues)
fig.colorbar(cax)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks([0, 1], ['BMI', 'PHYSACTIVITY'])
plt.yticks([0, 1], ['BMI', 'PHYSACTIVITY'])
```
