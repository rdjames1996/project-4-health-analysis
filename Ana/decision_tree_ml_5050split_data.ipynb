```python
# Import dependencies

import os
import pandas as pd
import numpy as np
```

```python
# Read in csv file
#file_path = pd.read_csv(f"/Users/thehazeplace/Desktop/uci/project-4-health-analysis/Resources/diabetes_binary_health_indicators_BRFSS2015.csv")

file_path = pd.read_csv(f"/Users/thehazeplace/Desktop/uci/project-4-health-analysis/Resources/diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
file_path.head(150)
```

```python
# Convert csv to dataframe
diabetes_df = pd.DataFrame(file_path)
diabetes_df
```

```python
# List columns for review
diabetes_df.columns.to_list()
```

```python
# Drop unnecessary columns
diabetes_df_drop = diabetes_df.drop(columns=['NoDocbcCost', 'GenHlth', 'MentHlth', 'PhysHlth', 'DiffWalk'])
diabetes_df_drop
```

```python
#diabetes_df_drop.rename(columns={'Diabetes_binary': 'Diabetes_Status'}, inplace=True)
```

```python
#Filter through df to pull needed columns and rename 'Diabetes_binary ' to 'Diabetes_status' to pull target variable and features

consumption_df = diabetes_df.filter(['Diabetes_binary', 'Smoker', 'HvyAlcoholConsump', 'Fruits', 'Veggies'], axis=1)
consumption_df.rename(columns={'Diabetes_binary': 'Diabetes_Status'}, inplace=True)
consumption_df
```

```python
# Randomize dataset prior to training

consumption_df_random = consumption_df.sample(frac=1, random_state=1)
consumption_df_random
```

```python
# Define features set

X = consumption_df_random.copy()
X.drop('Diabetes_Status', axis=1, inplace=True)
X.head()
```

```python
# Define target vector

y = consumption_df_random["Diabetes_Status"].values.reshape(-1, 1)
y[:5]
```

```python
from sklearn.model_selection import train_test_split

# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

# Creating StandardScaler instance
scaler = StandardScaler()
```

```python
# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)
```

```python
# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Fitting the Decision Tree Model

```python
from sklearn import tree

# Needed for decision tree visualization
import pydotplus
from IPython.display import Image
```

```python
# Creating the decision tree classifier instance
model = tree.DecisionTreeClassifier()
```

```python
# Fitting the model
model = model.fit(X_train_scaled, y_train)
```

## Making Predictions Using the Tree Model

```python
# Making predictions using the testing data
predictions = model.predict(X_test_scaled)
```

## Model Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

## Visualizing the Decision Tree

```python
# Create DOT data
dot_data = tree.export_graphviz(
    model, out_file=None, feature_names=X.columns, class_names=["0", "1"], filled=True
)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)

# Show graph
Image(graph.create_png())
```
