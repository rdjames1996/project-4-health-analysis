```python
# Read in csv file
file_path = pd.read_csv(f"/Users/thehazeplace/Desktop/uci/project-4-health-analysis/Resources/diabetes_binary_health_indicators_BRFSS2015.csv")
#file_path = pd.read_csv(f"/Users/thehazeplace/Desktop/uci/project-4-health-analysis/Resources/diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
file_path
```

```python
# Convert csv to dataframe
diabetes_df = pd.DataFrame(file_path)
diabetes_df
```

```python
# List columns for review
diabetes_df.columns.to_list()
```

```python
# Drop unnecessary columns
diabetes_df_drop = diabetes_df.drop(columns=['NoDocbcCost', 'GenHlth', 'MentHlth', 'PhysHlth', 'DiffWalk'])
diabetes_df_drop
```

```python
diabetes_df_drop.groupby("Diabetes_binary").count()
```

```python
# Filter through df to pull needed columns and rename 'Diabetes_binary ' to 'Diabetes_status' to pull target variable and features
consumption_df = diabetes_df.filter(['Diabetes_binary', 'Smoker', 'HvyAlcoholConsump', 'Fruits', 'Veggies'], axis=1)
consumption_df.rename(columns={'Diabetes_binary': 'Diabetes_Status'}, inplace=True)
consumption_df
```

```python
# Convert new dataframe to csv file for future refernces
consumption_df.to_csv("../Resources/consumption_df.csv", index=False)
```

```python
# Check df and data types  
consumption_df.dtypes
```

```python
# Randomize dataset prior to training

consumption_df_random = consumption_df.sample(frac=1, random_state=1)
```

```python
# Split data into target variable and features
X = consumption_df_random.drop(columns=['Diabetes_Status'], axis=1)
y = consumption_df_random['Diabetes_Status']
```

```python
X 
```

## Split data into training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state=1,
                                                    stratify=y)
```

## Create a Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

## Fit (train) or Model using the training data

```python
# Train the data
classifier.fit(X_train, y_train)
```

## Validating the model using the test data

```python
# Score the data
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Tesitng Data Score: {classifier.score(X_test, y_test)}")
```

## Make Predictions

```python
# Predict outcomes for test data set
predictions =classifier.predict(X_test)
predictions_df = pd.DataFrame({"Prediction": predictions, "Actual": y_test})
predictions_df
```

```python
predictions
```

```python
y_test
```

## Calculate the Accuracy Score

```python
from sklearn.metrics import accuracy_score
# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```

## Confusion Matrix & Classification Report

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
target_names = ["0", "1"]
print(classification_report(y_test, predictions, target_names=target_names))
```

Based on the confusion matrix, this model is proving to be inaccurate.
