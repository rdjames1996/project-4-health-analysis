```python
# Import the modules
import numpy as np
import pandas as pd
import os
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Convert csv to dataframe
diabetes_csv = r"C:\Users\stave\OneDrive\Desktop\Bootcamp\Project 4\Resources\50_50_split.csv"
diabetes_binary_df = pd.read_csv(diabetes_csv)
diabetes_binary_df
```

```python
# Drop unnecessary columns
diabetes_binary_df_clean = diabetes_binary_df.drop(columns=['NoDocbcCost', 'GenHlth', 'MentHlth', 'PhysHlth', 'DiffWalk', 'CholCheck'])
diabetes_binary_df_clean
```

```python
#Create dataframe with pre-existing conditions and diabetes status
preexisting_binary_df = diabetes_binary_df_clean.filter(['Diabetes_binary', 'HighBP', 'Age', 'BMI', 'Sex'])
preexisting_binary_df
```

```python
#Define X and y
y = diabetes_binary_df_clean['Diabetes_binary']
X = diabetes_binary_df_clean.drop(columns='Diabetes_binary')
```

```python
# Split the dataset using train_test_split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
#Create a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
 # Train the data

classifier.fit(X_train, y_train)
```

```python
 # Score the model
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Predict outcomes for test data set
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
#Calculate Accuracy Score
from sklearn.metrics import accuracy_score
# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```

```python
#Create confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
#Create Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
